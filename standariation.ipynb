```python
Scaler =  StandardScaler()
```

# formula apply

```python
x_train_s= Scaler.transform(x_train)
x_test_s= Scaler.transform(x_test)
```

```python
x_train_s
```

```python
x_train_st = pd.DataFrame(x_train_s,columns=x_train.columns)
x_test_st = pd.DataFrame(x_test_s,columns=x_test.columns)
```

# effet of saling

```python
fig, (a1,a2) = plt.subplots(ncols=2, figsize=(12,5))
a1.scatter(x_train['Age'],x_train['EstimatedSalary'])
a1.set_title('before')
a2.scatter(x_train_st['Age'],x_train_st['EstimatedSalary'])
a2.set_title('after')
```

```python
fig, (a1,a2) = plt.subplots(ncols=2, figsize=(12,5))
sns.kdeplot(x_train['Age'],ax=a1)
sns.kdeplot(x_train['EstimatedSalary'],ax=a1)

sns.kdeplot(x_train_st['Age'],ax=a2)
sns.kdeplot(x_train_st['EstimatedSalary'],ax=a2)
a2.set_title('after')
```
